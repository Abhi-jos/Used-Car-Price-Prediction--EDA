```python
from scipy import stats
Q1=df['selling_price'].quantile(0.25)
Q1
```

```python
Q3=df['selling_price'].quantile(0.75)
Q3
```

```python
IQR=Q3-Q1
IQR
```

```python
from scipy import stats
IQR = stats.iqr(df['selling_price'])
IQR
```

```python
lower_limit=Q1-1.5*IQR
upper_limit=Q3+1.5*IQR
lower_limit,upper_limit
```

```python
df[(df['selling_price']<lower_limit)|(df['selling_price']>upper_limit)]
```

```python
df_outliers=df[(df['selling_price']<lower_limit)|
               (df['selling_price']>upper_limit)]
```

```python
df_outliers.shape
```

```python
index_of_outliers=df[(df['selling_price']<lower_limit)|
         (df['selling_price']>upper_limit)].index
index_of_outliers
```

```python
df.drop(index_of_outliers, inplace=True)
```
